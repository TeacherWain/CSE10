# Introduction to Robotics and Robotic Programming

<h2>What is Robotics?</h2>



A robot is a mechanical device that can perform tasks and interact with its environment, without the aid of human interaction.

Robotics is the science and technology behind the design, manufacturing, and application of robotics.

<h2>Basic Components of a Robot</h2>

1. <b>Body / Frame:</b> Can be any size, this is the structure of the robot
2. <b>Control System:</b> Equivalent to the central nervous system of a human
3. <b>Manipulators:</b> Robots have to be able to interact with their environment, sometimes they have to move or reorient objects without human contact. These are not integral to the robot.
4. <b>Drivetrain:</b> Some robots stay in place, but for this course we will mainly be working with ones that move. Drivetrains consist of a powered method of mobility.

<h2>Famous Robots</h2>



<b>Opportunity</b> was one of two landers sent to Mars in 2003. Originally planned to only last 90 days, Opportunity spent 5111 sols (1 sol = 24 h 39 min), or a total of 15 years. It was declared dead in early 2019 after last being heard from June 2018.



<b>Curiosity</b> landed on Mars August 6, 2012. Its purpose was to find biological signatures, geological and geochemical composition, planetary information, and measure surface radiation. It continues its mission in the Gale Crater to this day.

<h2>Types of Robots</h2>

<h3>1. Industrial Robots</h3>

There are many jobs that require speed and precision. Industrial processes have been greatly sped up and performed with a higher degree of precision by the use of robotics. 

<b>Examples:</b> 
- Packaging
- Assembly
- Painting
- Palletizing

Initially, robots only did very specialized and repetitive jobs, but now we are seeing robots that are flexible and make complex decision based on sensor feedback.



<h2>2. Robots in Research</h2>



Most famously, NASA has sent rovers (Pathfinder, Sojourner, Spirit, Opportunity, and Curiosity) to Mars to explore and search for water. 

This is because robots are able to perform tasks and go places where humans cannot.
